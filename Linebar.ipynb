```python
import matplotlib.pyplot as plt
x=[0,1,2,3,4,5]
y=[0,10,20,30,40,50]
plt.plot(x,y)
plt.title("Line chart")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[0,1,2,3,4,5]
y=[i**2 for i in x ]
plt.plot(x,y)
plt.title("Line chart")
plt.xlabel("x-axis")
plt.ylabel("y-squares")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.arange(0,10,2)
y=[i**2 for i in x]
print(x,y)
plt.plot(x,y)
plt.title("Line chart")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.arange(0,10,2)
y=x**2
print(x,y)
plt.plot(x,y,marker="o",ms=15,mec="r",mfc="w")
plt.title(" marker Line chart")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.arange(0,10,2)
y=x**2
print(x,y)
plt.plot(x,y,ls="dashed",lw=5,c="r")
plt.title("Line style chart")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.arange(0,10,2)
y1=x**2
y2=x**3
print(x,y)
plt.plot(x,y1)
plt.plot(x,y2)
plt.title("multiple Line chart")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```
